```python
import pandas as pd

# Assuming 'Clusters' is the column containing cluster labels, and you want to find top 3 values for each column
cluster_counts_top3 = {}

# Loop through each cluster
for cluster_label, cluster_data in cf.groupby('Clusters'):
    # Drop non-categorical columns if needed
    cluster_data_categorical = cluster_data.select_dtypes(include=['object'])

    # Initialize a dictionary to store top 3 values for each column
    top3_values = {}

    # Loop through each categorical column
    for column in cluster_data_categorical.columns:
        # Count the occurrences of each unique value
        value_counts = cluster_data[column].value_counts()

        # Select the top 3 values
        top3 = value_counts.head(3).index.tolist()

        # Store the top 3 values in the dictionary
        top3_values[column] = top3

    # Store the top 3 values for the current cluster in the main dictionary
    cluster_counts_top3[cluster_label] = top3_values

# Convert the dictionary to a DataFrame for better display
result_df_top3 = pd.DataFrame.from_dict({(i, j): cluster_counts_top3[i][j]
                                          for i in cluster_counts_top3.keys()
                                          for j in cluster_counts_top3[i].keys()},
                                         orient='index')


# Display the result
print(result_df_top3)
```
