```python
import pandas as pd
```

```python
graves = pd.read_csv('graves.csv')
```

```python
graves.head()
```

```python
graves.columns
```

```python
grave_col_list = ['Site_ID', 'GID','Number_object_records', 'Number_hr_records']
graves = graves[grave_col_list]
graves.head()
```

```python
hr = pd.read_csv('Human Remains.csv')
hr.head()
```

```python
hr.columns
```

```python
l2 = ['GID', 'GID_hr', 'HR_type']
hr = hr[l2]
hr.head()
```

```python
obj = pd.read_csv('Objects.csv')
obj.head()
```

```python
obj.columns
```

```python
l3 = ['GID', 'Object_type','Materials_summary']
obj = obj[l3]
obj.head()
```

```python
obj_dates = pd.read_csv('Object_Dates.csv')
obj_dates.head()
```

```python
obj_dates.columns
```

```python
l4 = ['GID_date', 'Period']
obj_dates = obj_dates[l4]
obj_dates.head()
```

```python
obj_dates.rename(columns = {'GID_date':'GID'}, inplace = True)
obj_dates.head()
```

```python
sites = pd.read_csv('Sites.csv')
sites.head()
```

```python
sites.loc[sites['Case_study_area'].isnull(), 'Case_study_area'] = 'Orkney and Outer Hebrides'
```

```python
sites.columns
```

```python
l5 = ['Site_ID','Case_study_area']
sites = sites[l5]
sites.head()
```

```python
graves_hr = pd.merge(graves, hr, on='GID', how='left')
graves_hr.head()
```

```python
graves_hr_site = pd.merge(graves_hr, sites, on='Site_ID', how='left')
graves_hr_site.head()
```

```python
graves_hr_site.isnull().sum()
```

```python
graves_hr_site_obj = pd.merge(graves_hr_site, obj, on='GID', how='left')
graves_hr_site_obj.head()
```

```python
graves_hr_site_obj.isnull().sum()
```

```python
final_df = pd.merge(graves_hr_site_obj, obj_dates, on='GID', how='left')
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df.shape
```

```python
category_mapping = {
    'Animal Bone (Other)' : 'Animal Remains',
    'Pot (Unknown/Unspecified)' : 'Pot',
    'Brooch (Specified type)':'Brooch',
     'Bracelet/Bangle (Unknown/Unspecified)':'Bracelet',
 'Brooch (Unknown/Unspecified)' :'Brooch',
 'Tool (Unknown/Unspecified)': 'Tool',
 'Dagger (Unknown/Unspecified)':'Dagger',
'Axe/Axehead (Unknown/Unspecified)' :'Axe',
 'Animal Burial (Specified type)' : 'Animal Remains',
 'Vessel (Unknown/Unspecified)' : 'Pot',
    'Pin (Unknown/Unspecified)' :'Pin',
'Animal Joint/Part' : 'Animal Remains',
'Arrowhead (Unknown/Unspecified)' : 'Arrowhead',
'Chariot/Cart Fitting (Unknown/Unspecified/Set)':'Chariot/Cart',
 'Animal Joint/Part (Other specified part)'  : 'Animal Remains',
'Axe/Axehead':'Axe',
 'Pot ' :'Pot',
 'Tool (Specified type)' : 'Tool',
 'Pin (Specified type)' :'Pin',
 'Animal Joint/Part (Unknown/Unspecified)' : 'Animal Remains',
 'Bracelet/Bangle (Specified type)':'Bracelet'
}
final_df['Object_type'] = final_df['Object_type'].replace(category_mapping)
```

```python
final_df.isnull().sum()
```

```python
final_df.drop('Site_ID', inplace=True, axis=1)
final_df.head()
```

```python
final_df.shape
```

```python
final_df.columns
```

```python
final_df_grouped = final_df.groupby('GID').agg(
    Number_object_records=('Number_object_records', lambda x: list(set(x)) if len(set(x)) > 1 else x.iloc[0]),
    Number_hr_records=('Number_hr_records', lambda x: list(set(x)) if len(set(x)) > 1 else x.iloc[0]),
    GID_hr=('GID_hr', lambda x: list(set(x)) if len(set(x)) > 1 else x.iloc[0]),
    HR_type=('HR_type', lambda x: list(set(x)) if len(set(x)) > 1 else x.iloc[0]),
    Case_study_area=('Case_study_area', lambda x: list(set(x)) if len(set(x)) > 1 else x.iloc[0]),
    Object_type=('Object_type', lambda x: list(set(x)) if len(set(x)) > 1 else x.iloc[0]),
    Materials_summary=('Materials_summary', lambda x: list(set(x)) if len(set(x)) > 1 else x.iloc[0]),
    Period=('Period', lambda x: list(set(x)) if len(set(x)) > 1 else x.iloc[0])
)

# Resetting index to make 'GID' a column again
final_df_grouped.reset_index(inplace=True)
final_df_grouped.head()
```

```python
final_df_grouped.head(10)
```

```python
final_df_grouped.shape
```

```python
final_df_grouped.to_csv('final_graves_combined.csv',index=False)
```

```python
cornwall_df = df[df['Case_study_area'].str.contains('Cornwall', case=False, na=False)]
cornwall_df.shape
```

```python
cornwall_df.loc[8:13,]
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Assuming 'final_df' is your DataFrame
# Assuming 'Grave_ID' is another column in your DataFrame
# Assuming 'Objects' is the column with lists of objects
df_transactional = cornwall_df[['GID', 'Object_type']].dropna()

# Convert the lists of objects to a transactional format
te = TransactionEncoder()
te_ary = te.fit(df_transactional['Object_type']).transform(df_transactional['Object_type'])

# Create a new DataFrame with the transactional format
df_transactional_encoded = pd.DataFrame(te_ary, columns=te.columns_)
df_transactional_encoded['GID'] = df_transactional['GID']

# Apply the Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(df_transactional_encoded.drop(columns=['GID']), min_support=0.01, use_colnames=True)

# Display the frequent itemsets
print("Frequent Itemsets:")
print(frequent_itemsets)
```
